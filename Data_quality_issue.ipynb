## What is the data quality Issue in the given data?

```python
import json
import pprint as pp

users_raw = [] #We will open the users json file and store all the dictionaries in this list
jsonraw = open('users.json', 'r')

for ele in jsonraw:
    x = json.loads(ele)
    users_raw.append(x) # Stored individual dictionaires in the receipt_raw list
```

```python
for count,ele in enumerate(users_raw):
    if count == 5:
        break
    pp.pprint(ele) #printing first 5 dictionaries for a quick view of the list
```

```python
test = []
for ele in users_raw: # Creating list of unique UserIDs
    for key,value in ele.items():
        if key == '_id':
            tempdict = {value['$oid'] : 0}
            test.append(tempdict)

for ele in users_raw: # Checking how many times the userID is repeated in the database and storing it
    for key,value in ele.items():
        if key == '_id':
            for i in test:
                for key1,value1 in i.items():
                    if value['$oid'] == key1:
                        i[key1] = value1 +1
```

```python
pp.pprint(test) #printing the final output of the userIDs and the count of repetition of each userID
```

As we see above that many userIDs are repeated multiple times in the Users Dataset. Repetition of data is major issue in Analytical world
